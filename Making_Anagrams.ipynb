We consider two strings to be anagrams of each other if the first string's letters can be rearranged to form the second string. In other words, both strings must contain the same exact letters in the same exact frequency.



```
#Create a MakeAnagram function below.
def MakeAnagram(str1,str2):
  '''determine the minimum number of character deletions required to make and anagrams'''

  # Step 1. Getting a list which has unique elements only for both the strings

  list1 = [i for i in str1]
  set1 = set(list1)
  unique_list1  = list(set1)
  
  list2 = [i for i in str2]
  set2 = set(list2)
  unique_list2  = list(set2)
  

  # Step 2. Using a dictionary to store the unique elements and the number of repetition
  
  dict1= {}                           
  for i, j in enumerate (unique_list1):
    dict1[unique_list1[i]] = list1.count(j)
  print(dict1)

  dict2= {}
  for i, j in enumerate (unique_list2):
    dict2[unique_list2[i]] = list2.count(j)
  print(dict2)


  # Step 3. Counting the number to elements to delete for making the 2 strings Anagram

  num = 0
  for key1, value1 in dict1.items():
    if key1 not in dict2.keys():
       num += value1

  for key2, value2 in dict2.items():
    if key2 not in dict1.keys():
       num += value2
  
  for key1, value1 in dict1.items():
    for key2, value2 in dict2.items():
      if  key1 in dict2.keys():
        if key1 == key2:
          num += abs(value1-value2)
  
  print(f"The minimum number of character deletions required to make {str1}  and  {str2}  anagrams is {num}")
```

```
# Calling the function MakeAnagram() to check on the two strings
string1 = "bqcdd"
string2 = "ddeebb"
MakeAnagram(string1, string2)
```
